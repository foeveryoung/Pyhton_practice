```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
import csv
import seaborn

def load_data(path):
    files = os.listdir(path)
    data = pd.DataFrame()
    for file in  files[:]:
        if len(file)==8:
            tmp = pd.read_csv(path + file, sep='\u0003', error_bad_lines=False, header = None,quoting=csv.QUOTE_NONE)
            # provinceid 32
            # tmp = tmp[tmp[32] == 32]
            data= pd.concat([data, tmp], axis=0)
            print(file)
    return data
```

```python
def tel_check(x):
    return str(x).replace('-','').replace(' ',';').replace('、',';').replace('、 ',';').replace(r'/',';').replace(r'，,',';')
```

```python
ad4['financetel']=ad4['financetel'].apply(lambda x :tel_check(x))
ad4.loc[ad4['financetel'].apply(len)<7,'financetel']=np.nan
```

```python
ad4['financemotel']=ad4['financemotel'].apply(lambda x :tel_check(x))
ad4.loc[ad4['financemotel'].apply(len)<7,'financemotel']=np.nan
```
